```python
# Import libraries

import pandas as pd
import numpy as np
import os
import gc
```

```python
# Defining path
path = r'C:\Users\nukis\Documents\Instacart Basket Analysis'
```

```python
# Importing resized version of 'orders_products_merged_update2.pkl'

df_ords_prods_cleaned = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_update2_resized.pkl'))
```

```python
# Import again the cleaned version of 'df_cust_clean'

df_cust_cleaned = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'customer_cleaned.pkl'))
```

```python
df_cust_cleaned.info(memory_usage='deep')
```

```python
# Looking at the size of this DataFrame, I think, reducing 'object' datatype is not really that relevant.

df_cust_cleaned_obj = df_cust_cleaned.select_dtypes(include=['object']).copy()
df_cust_cleaned_obj.describe()
```

```python
# Change in datatype by downcasting loop

for column in df_cust_cleaned:
    if df_cust_cleaned[column].dtype == 'int64':
        df_cust_cleaned[column] = pd.to_numeric(df_cust_cleaned[column], downcast = 'integer')
```

```python
df_cust_cleaned.info(memory_usage='deep')
```

```python
df_cust_cleaned.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'customer_cleaned_resized.pkl'))
```

```python
# Merging resized 'df_ords_prods_clean' and 'df_cust_clean'. The data type 'user_id' of both DataFrame is already the same.

df_ords_prods_cust_cleaned = df_ords_prods_cleaned.merge(df_cust_cleaned, on = 'user_id', indicator = True)
```

```python
# Export the merged 'df_ords_prods_cleaned' and 'df_cust_cleaned'

df_ords_prods_cust_cleaned.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_cust_cleaned.pkl'))
```
